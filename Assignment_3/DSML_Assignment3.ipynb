```python
#Apply appropriate ML algorithm on a dataset collected in a cosmetics shop showing details of customers to predict customer response for special offers.
```

```python
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv("/content/Mall_Customers.csv")
```

```python
ds.head()
```

```python
ds = pd.get_dummies(ds, columns=["Genre"])
```

```python
ds.head()
```

```python
X = ds.iloc[:, [0,1,2,4,5]]
y = ds.iloc[:, [3]]
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, y_Train, y_Test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = "linear", random_state = 0)
```

```python
classifier.fit(X_Train, y_Train)
```

```python
y_pred = classifier.predict(X_Test)
```

```python
from sklearn import metrics
```

```python
print("Mean absolute error: ", metrics.mean_absolute_error(y_Test, y_pred))
print("Mean squared error: ", metrics.mean_squared_error(y_Test, y_pred))
print("R2 score: ", metrics.r2_score(y_Test, y_pred))
```
